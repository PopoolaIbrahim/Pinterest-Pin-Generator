```python
from PIL import Image, ImageDraw, ImageFont
import random

keyword = "Summer Nails"
########## Resizing Product image
img = Image.open("/content/1.jpg")
new_width = 680
new_height = 900
img = img.resize((new_width, new_height), resample=Image.BICUBIC)
img.save("/content/2.jpg")
##### Create Shop Now image
img1 = Image.new('RGB', (200, 40), color="red")
draw = ImageDraw.Draw(img1)
font = ImageFont.truetype("/content/font2.ttf", 33)
draw.text((5, 0), "SHOP NOW", font=font, fill='white')
img1.save("/content/4.png")
##### Creating Background
color_chosen = random.choice(["brown", "purple", "#0097b2", "blue", "#10215B", "#5B1048", "#CD0B3F", "#84641C", "#9D1612"])
img1 = Image.new('RGB', (680, 1020), color=color_chosen)
draw = ImageDraw.Draw(img1)
font = ImageFont.truetype("/content/font1.ttf", 60)
draw.text((150, 0), keyword, font=font, fill='white')
img1.save("/content/5.png")
##### Creating Pin
post_image = "/content/6.jpg"
img1 = Image.open("/content/5.png")
img2 = Image.open("/content/2.jpg")
img4 = Image.open("/content/4.png")
img1.paste(img2, (0, 120))
img1.paste(img4, (240, 900))
img1.save(post_image, quality=95)
display(img1)

```
